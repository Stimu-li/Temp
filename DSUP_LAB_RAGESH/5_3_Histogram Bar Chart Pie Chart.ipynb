```python
import matplotlib.pyplot as plt

# --- Histogram ---
data = [0.5, 1, 1.5, 2, 3, 4, 5, 6, 8, 9]
plt.hist(data, bins=4, color="skyblue", edgecolor="black")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()

# --- Bar Chart ---
languages = ["JavaScript", "Java", "Python", "C#"]
usage = [70, 45, 39, 34]
plt.bar(languages, usage, color="orange")
plt.ylabel("Usage (%)")
plt.title("Programming Language Usage")
plt.show()

# --- Pie Chart ---
animals = ["Cats", "Dogs", "Tigers", "Goats"]
sizes = [25, 20, 45, 10]
plt.pie(sizes, labels=animals, autopct="%.1f%%")
plt.title("Animal Distribution")
plt.show()
```
