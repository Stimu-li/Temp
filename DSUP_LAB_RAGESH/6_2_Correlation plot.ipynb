```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Iris-like data
data = {
    "SepalLengthCm": [5.1, 4.9, 6.2, 5.5, 5.8],
    "SepalWidthCm": [3.5, 3.0, 2.8, 4.2, 3.9],
    "Species": ["setosa", "setosa", "versicolor", "versicolor", "virginica"],
}
df = pd.DataFrame(data)

# Scatter plot by Species
sns.FacetGrid(df, hue="Species", height=5).map(
    plt.scatter, "SepalLengthCm", "SepalWidthCm"
).add_legend()
plt.show()

# Correlation matrix heatmap
corr = df[["SepalLengthCm", "SepalWidthCm"]].corr()
sns.heatmap(corr, annot=True)
plt.show()
```
