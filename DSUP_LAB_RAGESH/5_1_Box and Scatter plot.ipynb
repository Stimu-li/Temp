```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Scatter Plot ---
data = pd.DataFrame(
    {
        "SepalLengthCm": [5.1, 4.9, 6.2, 5.5, 5.8],
        "SepalWidthCm": [3.5, 3.0, 2.8, 4.2, 3.9],
    }
)
plt.scatter(data["SepalLengthCm"], data["SepalWidthCm"], color="blue")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot - Iris Sample")
plt.show()

# --- Box Plot ---
np.random.seed(10)
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)

plt.boxplot([data_1, data_2, data_3, data_4])
plt.title("Box Plot of Random Data")
plt.xlabel("Datasets")
plt.ylabel("Values")
plt.show()
```
